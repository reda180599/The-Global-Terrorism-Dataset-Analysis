The "Global Terrorism" dataset is a comprehensive dataset on terrorist attacks worldwide from 1970 to 2017. It contains information on over 180,000 terrorist incidents, including details such as the date and location of the attack, the type of attack, the target, the group responsible, and the number of fatalities and injuries.

The dataset contains the following columns:

Year: The year in which the terrorist incident occurred.
Month: The month in which the terrorist incident occurred.
Day: The day of the month on which the terrorist incident occurred.
Country: The country in which the terrorist incident occurred.
Region: The region in which the terrorist incident occurred.
City: The city in which the terrorist incident occurred.
Latitude: The latitude of the location in which the terrorist incident occurred.
Longitude: The longitude of the location in which the terrorist incident occurred.
AttackType: The type of terrorist attack, such as bombing or assassination.
TargetType: The type of target of the terrorist attack, such as government or business.
TargetSubtype: A more specific description of the target type.
TargetNationality: The nationality of the target of the terrorist attack.
GroupName: The name of the group responsible for the terrorist attack.
GroupSubname: A more specific description of the group responsible.
GroupMotive: The motive of the group responsible.
WeaponType: The type of weapon used in the terrorist attack.
WeaponSubtype: A more specific description of the weapon type.
Fatalities: The number of fatalities in the terrorist attack.
Injuries: The number of injuries in the terrorist attack.
TotalCasualties: The total number of casualties (fatalities + injuries) in the terrorist attack.
Target: A more detailed description of the target.
Summary: A summary of the terrorist attack.

```python
# Import required libraries
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
data = pd.read_csv('globalterrorismdb_0718dist.csv', encoding='ISO-8859-1', low_memory=False)

```

```python
# Display the first 5 rows of the data
print(data.head())
```

```python
# Display the shape of the data
print("Data shape:",data.shape)
```

```python
# Display the column names and datatypes
print("Column names and datatypes:",data.dtypes)
```

```python
# Summary statistics of numerical variables
print("Summary statistics of numerical variables:\n", data.describe())
```

```python
# Missing value analysis
print("Missing value analysis:\n", data.isnull().sum())
```

```python
# Remove rows with missing values of important variables
data = data.dropna(subset=['iyear', 'imonth', 'iday', 'country_txt', 'region_txt', 'city', 'latitude', 'longitude', 'attacktype1_txt', 'targtype1_txt', 'gname'])
```

```python
# Duplicate record analysis
print("Duplicate record analysis:", data.duplicated().sum())
```

```python
# Outlier analysis
fig, ax = plt.subplots(figsize=(10,5))
sns.boxplot(x=data['nkill'], ax=ax)
plt.show()
```

```python
# Impute missing values of selected variables
data['nkill'].fillna(data['nkill'].median(), inplace=True)
```

```python
# Feature engineering
data['casualty'] = data['nkill'] + data['nwound']
```

```python
# Hot zone of terrorism
hot_zones = data.groupby('country_txt')['casualty'].sum().reset_index().sort_values(by='casualty', ascending=False).head(10)
print("Hot zones of terrorism:\n", hot_zones)
```

```python
# Analyze the distribution of terrorist attacks by year
attacks_by_year = data['iyear'].value_counts().sort_index()
fig, ax = plt.subplots(figsize=(10,5))
ax.plot(attacks_by_year.index, attacks_by_year.values)
ax.set_xlabel('Year')
ax.set_ylabel('Number of attacks')
ax.set_title('Terrorist attacks by year')
plt.show()
```

```python
# Analyze the distribution of terrorist attacks by region
attacks_by_region = data['region_txt'].value_counts()
fig, ax = plt.subplots(figsize=(10,5))
sns.barplot(x=attacks_by_region.index, y=attacks_by_region.values, ax=ax)
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
ax.set_xlabel('Region')
ax.set_ylabel('Number of attacks')
ax.set_title('Terrorist attacks by region')
plt.show()
```

```python
# Analyze the distribution of terrorist attacks by attack type
attack_types = data['attacktype1_txt'].value_counts()
fig, ax = plt.subplots(figsize=(10,5))
sns.barplot(x=attack_types.index, y=attack_types.values, ax=ax)
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
ax.set_xlabel('Attack type')
ax.set_ylabel('Number of attacks')
ax.set_title('Types of terrorist attacks')
plt.show()
```

```python
# Analyze the distribution of terrorist attacks by target type
target_types = data['targtype1_txt'].value_counts()
fig, ax = plt.subplots(figsize=(10,5))
sns.barplot(x=target_types.index, y=target_types.values, ax=ax)
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
ax.set_xlabel('Target type')
ax.set_ylabel('Number of attacks')
ax.set_title('Targets of terrorist attacks')
plt.show()
```

```python
# Analyze the distribution of terrorist attacks by weapon type
weapon_types = data['weaptype1_txt'].value_counts()
fig, ax = plt.subplots(figsize=(10,5))
sns.barplot(x=weapon_types.index, y=weapon_types.values, ax=ax)
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
ax.set_xlabel('Weapon type')
ax.set_ylabel('Number of attacks')
ax.set_title('Types of weapons used in terrorist attacks')
plt.show()
```

```python
# Analyze the distribution of terrorist attacks by group name
group_names = data['gname'].value_counts().head(10)
fig, ax = plt.subplots(figsize=(10,5))
sns.barplot(x=group_names.index, y=group_names.values, ax=ax)
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
ax.set_xlabel('Group name')
ax.set_ylabel('Number of attacks')
ax.set_title('Terrorist groups with the most attacks')
plt.show()
```

```python
# Analyze the distribution of terrorist attacks by group name
group_names = data['gname'].value_counts().head(10)
fig, ax = plt.subplots(figsize=(10,5))
sns.barplot(x=group_names.index, y=group_names.values, ax=ax)
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
ax.set_xlabel('Group name')
ax.set_ylabel('Number of attacks')
ax.set_title('Terrorist groups with the most attacks')
plt.show()
```

```python
# Heatmap of terrorist attacks by region and weapon type
region_weapontype = pd.crosstab(index=data['region_txt'], columns=data['weaptype1_txt'])
plt.figure(figsize=(10,8))
sns.heatmap(region_weapontype, annot=True, cmap='Blues')
plt.xlabel('Weapon type')
plt.ylabel('Region')
plt.title('Terrorist attacks by region and weapon type')
plt.show()
```
